```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
```

```python
data= pd.read_csv("health care diabetes.csv")
```

```python
data.info()      
df= data.copy()
```

there are no missing values in this dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Glucose Distribution
axes[0].hist(df['Glucose'], bins=10)
axes[0].set_xlabel('Glucose')
axes[0].set_ylabel('Count')
axes[0].set_title('Histogram of Glucose')

# BMI Distribution
axes[1].hist(df['BMI'], bins=10)
axes[1].set_xlabel('BMI')
axes[1].set_ylabel('Count')
axes[1].set_title('Histogram of BMI')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Set up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Glucose Box Plot Grouped by Outcome
sns.boxplot(x='Outcome', y='Glucose', data=df, ax=axes[0])
axes[0].set_xlabel('Outcome')
axes[0].set_ylabel('Glucose')
axes[0].set_title('Box Plot of Glucose Grouped by Outcome')

# BMI Box Plot Grouped by Outcome
sns.boxplot(x='Outcome', y='BMI', data=df, ax=axes[1])
axes[1].set_xlabel('Outcome')
axes[1].set_ylabel('BMI')
axes[1].set_title('Box Plot of BMI Grouped by Outcome')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Set up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Glucose vs. BMI
axes[0].scatter(df['Glucose'], df['BMI'])
axes[0].set_xlabel('Glucose')
axes[0].set_ylabel('BMI')
axes[0].set_title('Scatter Plot of Glucose vs. BMI')

# Age vs. Insulin
axes[1].scatter(df['Age'], df['Insulin'])
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Insulin')
axes[1].set_title('Scatter Plot of Age vs. Insulin')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Outcome Distribution (Diabetes vs. Non-Diabetes)
outcome_counts = df['Outcome'].value_counts()
plt.bar(outcome_counts.index, outcome_counts.values)
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Bar Plot of Outcome')
plt.xticks([0, 1], ['Non-Diabetes', 'Diabetes'])
plt.show()
```

```python
import numpy as np

# Calculate correlation matrix
corr_matrix = df.corr()

# Generate heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
